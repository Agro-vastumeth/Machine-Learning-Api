```python
import pandas as pd
import numpy as np
from pprint import pprint
import os
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import mean_squared_error

def evaluate(model, X_test, y_test):
    predictions = model.predict(X_test)
    errors = np.sqrt(mean_squared_error(y_test, predictions))
    print('Model Performance')
    print('MSE of: ', errors)
    
    return errors
```

```python
df = pd.read_csv('elemental.csv')
```

```python
df.head()
```

```python
df.dropna(axis=1, how="all")
```

```python
A = df.iloc[:,1:6]
B = df["Experimental (L/Kg)"]
frames = [A,B]
Reqdf = pd.concat(frames, axis=1)
Reqdf
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
# Reqdf["Ash"] = impute.fit_transform(Reqdf[["Ash"]]).ravel()
```

```python
X = Reqdf.iloc[:,:-1].values
Y = Reqdf.iloc[:,5].values

# Expected 2D array, got 1D array instead:
Y=Y.reshape((-1,1))
```

```python
X_train, X_test , Y_train , Y_test = train_test_split(X, Y, test_size=1, random_state=1)
```

```python
clf = RandomForestRegressor(random_state=1)
clf.fit(X_train, Y_train)
```

```python
base_accuracy = evaluate(clf,X_test, Y_test)
# print(X_test)
# y_pred = clf.predict([[20,30,20,10,10]])
```

```python
print('Parameters currently in use:\n')
pprint(clf.get_params())
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 1, stop = 50, num = 1)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(1, 30, num = 1)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = np.arange(1,20,1)
# Minimum number of samples required at each leaf node
min_samples_leaf = np.arange(1,20,1)
# restrict variability in the data selection
random_state = [1]
# Method of selecting samples for training each tree
# bootstrap = [True, False]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'random_state': random_state
               }

pprint(random_grid)
```

```python


best_random = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=30,
                                    max_features='sqrt', max_leaf_nodes=None,
                                    min_impurity_decrease=0.0,
                                    min_samples_leaf=2, min_samples_split=5,
                                    min_weight_fraction_leaf=0.0, n_estimators=200, n_jobs=1,
                                    oob_score=False, verbose=0, warm_start=False, random_state=1)
pprint(best_random.get_params())
best_random.fit(X_train, Y_train)
```

```python
random_accuracy = evaluate(best_random, X_test, Y_test)
```

```python
print('Improvement of {:0.2f}%.'.format((random_accuracy - base_accuracy) / base_accuracy))
```

```python
from sklearn.model_selection import GridSearchCV

grid_model = GridSearchCV(RandomForestRegressor(), random_grid, refit=True, verbose=0, cv=2)
grid_model.fit(X_train, Y_train)
grid_model.predict([[20,30,20,10,10]] )
```

```python
Grid_accuracy = evaluate(grid_model, X_test, Y_test)
print(Grid_accuracy)
```

```python
print('Improvement of {:0.2f}%.'.format((Grid_accuracy - random_accuracy) / random_accuracy))
```
